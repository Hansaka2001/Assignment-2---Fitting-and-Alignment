```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img_path = '/content/earrings.jpg'
img = cv2.imread(img_path)

if img is None:
    print(f"Error: Could not load image at {img_path}. Please upload it.")
else:
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.figure(figsize=(10, 10))
    plt.imshow(img_rgb)
    plt.title("Use the axes to estimate the width and height of the earrings in pixels")
    plt.grid(color='white', linestyle='--', linewidth=0.5, alpha=0.7)
    plt.xlabel("Width (Pixels)")
    plt.ylabel("Height (Pixels)")
    plt.show()
    plt.savefig('Q2_output.png', dpi=300, bbox_inches='tight')


    pixel_width = 150
    pixel_height = 200

    f = 8.0
    Z = 720.0
    pixel_size_mm = 2.2 * (10**-3)

    magnification_scale = Z / f

    real_width_mm = pixel_width * pixel_size_mm * magnification_scale
    real_height_mm = pixel_height * pixel_size_mm * magnification_scale

    print("\Results")
    print(f"Scale Factor (Real world / Sensor): {magnification_scale}")
    print(f"Calculated Real Width:  {real_width_mm:.2f} mm")
    print(f"Calculated Real Height: {real_height_mm:.2f} mm")
```
